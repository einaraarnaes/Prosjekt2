## Oppgave 1: Diffusjonslikningen

#### Oppgave 1a)

Vi ønsker å vise at likningen
$$\phi(x,t) = \frac{1}{\sqrt{4\pi Dt}} \int_{-\infty}^{\infty} \mathrm{d}y \, h(y) e^{-\frac{(x-y)^2}{4Dt}}$$

er en løsning at diffusjonslikningen i 1D, der D er en konstant:

$$\frac{\partial \phi(x,t)}{\partial t} = D \frac{\partial^2 \phi(x, t)}{\partial x^2}$$
Flytter først $t^{-1/2}$ inn i integralet:
$$\phi(x,t) = \frac{1}{\sqrt{4\pi D}} \int_{-\infty}^{\infty} \mathrm{d}y \, h(y) t^{-1/2}e^{-\frac{(x-y)^2}{4Dt}}$$

Antar at vi kan bytte om på integrasjons- og derivasjonsrekkefølgen. Siden det bare er den siste delen av integralet som avhenger av x og t trenger vi bare å vise at uttrykket $f(x,t) = t^{-1/2}e^{-\frac{(x-y)^2}{4Dt}}$ oppfyller likningen.
Regner først ut venstresiden:

$$\frac{\partial f}{\partial t} = -\frac{1}{2} t^{-3/2} e^{-\frac{(x-y)^2}{4Dt}} + t^{-1/2}\frac{(x-y)^2}{4Dt^2}e^{-\frac{(x-y)^2}{4Dt}}
= \left[ \frac{(x-y)^2}{4Dt} - \frac{1}{2} \right] t^{-3/2} e^{-\frac{(x-y)^2}{4Dt}}
$$

Deriverer to ganger med hensyn på x på høyresiden:
$$\frac{\partial f}{\partial x} = t^{-1/2} \frac{-2(x-y)}{4Dt} e^{-\frac{(x-y)^2}{4Dt}} = -\frac{1}{2D} t^{-3/2} \left[ (x-y) e^{-\frac{(x-y)^2}{4Dt}} \right]$$

$$\frac{\partial^2 f}{\partial x^2} = -\frac{1}{2D} t^{-3/2} \left[ e^{-\frac{(x-y)^2}{4Dt}} - \frac{2(x-y)^2}{4Dt} e^{-\frac{(x-y)^2}{4Dt}}\right] = \frac{1}{D} \left[ \frac{(x-y)^2}{4Dt} - \frac{1}{2} \right] t^{-3/2} e^{-\frac{(x-y)^2}{4Dt}}$$

Av dette ser vi lett at $\frac{\partial f}{\partial t} = D\frac{\partial^2 f}{\partial x^2}$, som var det vi ville vise.

Fysisk tolkning:

Hvis vi lar $t \rightarrow 0$ i $\phi(x,t)$ ser vi at eksponensialleddet i integralet går mot 0 i alle andre punkter enn $x = y$. Vi kan tolke dette som at alle partiklene starter samlet i punktet $x = y$ ved $t = 0$. Det må bety at funksjonen $h(y)$ på et vis beskriver hvordan spredningen av partiklene skjer avhengig av startposisjonen.

Fra formelen for integralet til en generell gaussfunksjon $\int_{-\infty}^{\infty} e^{-a(x+b)^2} \mathrm{d}x = \sqrt{\frac{\pi}{a}}$ Ser vi at funksjonen vår forenkler seg til et normert integral med verdi konstant lik C når $h(y) = C$. For funksjoner som avhenger av y vil vi få mer kompliserte fordelinger.

**Vent dette er fakka:** Vi integrerer jo bort y, så fordelingen er ikke avhengig av hvor vi starter? Men det vil jo det? Vi kan jo ikke evaluere integralet før vi har en y og en t. Er jo derfor vi har integralet der i utgangspunktet. Vi kan ikke evaluere det før vi har en x og en t.

#### Oppgave 1b)

Vi har at $P(x) = \frac{e^{-\beta V(x)}}{Z}$, der $\beta = (k_BT)^{-1}$. Vi lar sannsynligheten for at an partikkel beveger seg fra en posisjon x til $x_0 \in \{x-1, x, x+1\}$ er proposjonal med $P(x)$, og vi antar at denne proposjonalitetskonstanten er lik for alle mulige steg. VIl bestemme sannsynlighetene $p^+, p^0, p^-$ for at partikkelen flytter seg til høyre, blir eller til venstre. 

Vi har først kravet om at summen av de tre sannsynligghetene skal være 1 (vi kan se bort fra proposjonalitetsfaktoren).

$$\frac{e^{-\beta V(x_0 + 1)}}{Z} + \frac{e^{-\beta V(x_0)}}{Z} + \frac{e^{-\beta V(x_0 - 1)}}{Z} = 1$$
$$Z = e^{-\beta V(x_0 + 1)} + e^{-\beta V(x_0)} + e^{-\beta V(x_0 - 1)}$$

Vi får dermed:

$$ p^+ = P(x_0 + 1) = \frac{e^{-\beta V(x_0 + 1)}} {e^{-\beta V(x_0 + 1)} + e^{-\beta V(x_0)} + e^{-\beta V(x_0 - 1)}} = \frac{1}{1 + e^{-\beta[V(x_0-1)-V(x_0+1)]} + e^{-\beta[V(x_0)-V(x_0+1)]}}$$

Uttrykk for $p^0$ og $p^-$ får vi helt tilsvarende ved å regne ut $P(x_0)$ og $P(x_0-1)$, så vi skriver det ikke opp her.


#### Oppgave 1c) 

Vi ser på de to grensetilfellene for $p^+$, $p^0$ og $p^-$.

Når $k_{\mathrm{B}} T \gg |V(x+1)-V(x)|$ for alle x vil $\beta |V(x+1)-V(x)| \ll 1$. Det vil si at alle eksponensialleddene går mot $e^0 = 1$, og dermed får vi at $p^+ \approx p^0 \approx p^- \approx \frac{1}{3}$

Når $k_{\mathrm{B}} T \ll |V(x+1)-V(x)|$ for alle x vil $\beta |V(x+1)-V(x)| \gg 1$ det vil si at alle eksponensialleddene i $p^-$ går mot 0, slik at $p^- \approx 1$, mens $p_0$ og $p^+$ begge har eksponensialledd som vokser, slik at $p_0 \rightarrow 0$ og $p^+ \rightarrow 0$.

Dette synes å være fornuftig, små potensialforskjeller relativt til partiklenes energi ikke burde ha noen påvirkning på deres bevegelse, mens store potensialforskjeller burde gjøre det mer sannsynlig å bevege seg til lavere poteniale. Vi kan i disse to tilfellene forenkle virrevandringsproblemet, ved i det første å neglisjere potensialforskjellene slik at virrevandringen blir helt tilfeldig, og i det andre tilnærme det neste steget som helt sikkert. 



## Oppgave 2: Virrevandring

#### Oppgave 2a)

```python
import numpy as np
from matplotlib import pyplot as plt

def V_1(x):
    return 1

def V_2(x):
    return -x

def p_pluss_minus(x, V, bk):
    p_pluss = 1 / (1 + np.exp(-bk*(V(x-1)-V(x+1))) + np.exp(-bk*(V(x)-V(x+1))))
    p_minus = 1 / (1 + np.exp(-bk*(V(x+1)-V(x-1))) + np.exp(-bk*(V(x)-V(x-1))))
    return p_pluss, p_minus


"""
Funksjon som gjør numpsteps antall steg for hver posisjon i x,
og returnerer en array med endelig posisjon for hver partikkel.
"""
def random_walk(n_steps, x, V, bk):

    pos = np.copy(x)

    for i in range(n_steps):
        p_pluss, p_minus = p_pluss_minus(pos, V, bk)
        step = np.random.uniform(size = len(x))
        step[step <= p_minus] = -1
        step[(step > p_minus) & (step <= 1-p_pluss)] = 0
        step[step > 1-p_pluss] = 1

        pos = pos + step
    
    return pos


steps = 200
n_sim = 1
bk = 1

positions= np.arange(-50, 51, 1)
n_particles = len(positions)

walk_1 = np.zeros(n_particles)
walk_2 = np.zeros(n_particles)

for i in range(n_sim):
    walk_1 += random_walk(steps, positions, V_1, bk)
    walk_2 += random_walk(steps, positions, V_2, bk)

walk_1 /= n_sim
walk_2 /= n_sim

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 2.5))

axs[0].hist(walk_1, bins = 100)
axs[0].set_title("V(x)=k")
axs[1].hist(walk_2, bins = 100)
axs[1].set_title("V(x)=-kx")



for ax in axs:
    ax.set_xlabel("Position")
    ax.set_ylabel("Number of particles")
    #ax.legend()
```

#### Oppgave 2b)

```python

"""
Funksjon som gjør n_steps antall steg av random walk. 
En partikkel beveger seg en posisjon til høyre eller venstre med sannsynlighetene p_pluss og p_minus. 
Hvis en partikkel allerede befinner seg på posisjonen partikkelen ønsker å bevege seg, blir partikkelen værende.
Rekkefølgen på partiklene er i rekkefølgen til posisjonsvektoren.
"""
def random_walk_repulsion(n_steps, x, V, bk):

    pos_vector = np.copy(x)

    for i in range(n_steps):

        for i, pos in enumerate(pos_vector):
            p_pluss, p_minus = p_pluss_minus(pos, V, bk)
            step = np.random.uniform()

            if step <= p_minus:
                step = -1
            elif step > (1 - p_pluss):
                step = 1
            else:
                step = 0

            if not (pos_vector == (pos + step)).any():
                pos_vector[i] += step
    
    return pos_vector


walk_1_hard = np.zeros(n_particles)
walk_2_hard = np.zeros(n_particles)

for i in range(n_sim):
    walk_1_hard += random_walk_repulsion(steps, positions, V_1, bk)
    walk_2_hard += random_walk_repulsion(steps, positions, V_2, bk)

walk_1_hard /= n_sim
walk_2_hard /= n_sim

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 2.5))

axs[0].hist(walk_1_hard, bins = 200)
axs[0].set_title("V(x)=k")
axs[1].hist(walk_2_hard, bins = 200)
axs[1].set_title("V(x)=-kx")



for ax in axs:
    ax.set_xlabel("Position")
    ax.set_ylabel("Number of particles")
```

#### Oppgave 2c)

```python
bk = 1000

walk_1 = np.zeros(n_particles)
walk_2 = np.zeros(n_particles)
walk_1_hard = np.zeros(n_particles)
walk_2_hard = np.zeros(n_particles)

for i in range(n_sim):
    walk_1 += random_walk(steps, positions, V_1, bk)
    walk_2 += random_walk(steps, positions, V_2, bk)
    walk_1_hard += random_walk_repulsion(steps, positions, V_1, bk)
    walk_2_hard += random_walk_repulsion(steps, positions, V_2, bk)

walk_1 /= n_sim
walk_2 /= n_sim
walk_1_hard /= n_sim
walk_2_hard /= n_sim

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

axs[0,0].hist(walk_1, bins = 200)
axs[0,0].set_title("V(x)=k")
axs[0,1].hist(walk_2, bins = 200)
axs[0,1].set_title("V(x)=-kx")
axs[1,0].hist(walk_1_hard, bins = 200)
axs[1,0].set_title("V(x)=k, hard repulsion")
axs[1,1].hist(walk_2_hard, bins = 200)
axs[1,1].set_title("V(x)=-kx, hard repulsion")

for ax in axs.flatten():
    ax.set_xlabel("Position")
    ax.set_ylabel("Number of particles")


```

Kommentarer til oppgave 2:

Vi observerer i deloppgave a og b at med konstant potensial blir partiklene på samme område, mens de forflytter seg mot høyere når potensialet er synkende mot høyere. For kreteriet med "hard repulsjon" ser vi at partiklene har liknende oppførsel, men at spredningen blir større (siden partiklene ikke kan passere forbi hverandre). En måte å tenke på det er atomer i en gass som kolliderer med hverandre, slik at det oppstår et indre "trykk" som presser partiklene utover.

Videre ser vi at når vi senker temperaturen i deloppgave c, påvirker ikke dette partiklene under konstant potenial i særlig grad. For partiklene i det synkende potensialet derimot, blir bevegelsen til høyere enda mer dominerende. I tillegg blir forskjellen i spredning med og uten hard repulsjon enda tydeligere. Partiklene lengst til høyre beveger seg omtrent like langt som de som ikke er utsatt for hard repulsjon, men partiklene lengst til venstre har beveget seg mye kortere. En analogi til dette er biler som står i kø ved et trafikklys. Å ikke ha hard repulsjon blir som at alle sjåførene reagerer og begynner å kjøre i samme hastighet samtidig. Hard repulsjon blir nærmere slik det er i virkeligheten: Man kan ikke kjøre når det er en bil bare en mater foran. I tillegg har ikke alle sjåførene samme reaksjonstid, og resultatet er at bilkolonnen strekker seg ut.

## Oppgave 3: Virrevandring i skrallepotensial uten interaksjoner

#### Oppgave 3a)

```python
alpha = 0.8
T_p = 500
h = 1
N_x = 100
N_p = 12*N_x
bk = 1000
cycles = 10

def V_1(x):
    return np.where(x <= alpha*N_x, x/(alpha*N_x), x/((1-alpha)*N_x))

def V_2(x):
    return 1


positions = np.repeat(np.arange(0, 2*N_x, 1), N_p/(2*N_x))

currents = np.zeros(cycles)

for i in range(cycles):
    J_sum = 0
    for j in range(2*N_x):
        prev = positions
        if j < N_x:
            positions = random_walk(1, positions, V_1, bk)
        else:
            positions = random_walk(1, positions, V_2, bk)

        J_sum += np.sum(positions-prev)

        positions = positions%(2*N_x)
    
    currents[i] = J_sum/(N_p*2*N_x)
    

with np.printoptions(precision=5, suppress=True, linewidth = 100):
    print("Average currents for the 10 cycles:")
    print(currents)
```
