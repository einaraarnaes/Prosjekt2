## Oppgave 1: Diffusjonslikningen

#### Oppgave 1a)

Vi ønsker å vise at likningen
$$\phi(x,t) = \frac{1}{\sqrt{4\pi Dt}} \int_{-\infty}^{\infty} \mathrm{d}y \, h(y) e^{-\frac{(x-y)^2}{4Dt}}$$

er en løsning at diffusjonslikningen i 1D, der D er en konstant:

$$\frac{\partial \phi(x,t)}{\partial t} = D \frac{\partial^2 \phi(x, t)}{\partial x^2}$$
Flytter først $t^{-1/2}$ inn i integralet:
$$\phi(x,t) = \frac{1}{\sqrt{4\pi D}} \int_{-\infty}^{\infty} \mathrm{d}y \, h(y) t^{-1/2}e^{-\frac{(x-y)^2}{4Dt}}$$

Antar at vi kan bytte om på integrasjons- og derivasjonsrekkefølgen. Siden det bare er den siste delen av integralet som avhenger av x og t trenger vi bare å vise at uttrykket $f(x,t) = t^{-1/2}e^{-\frac{(x-y)^2}{4Dt}}$ oppfyller likningen.
Regner først ut venstresiden:

$$\frac{\partial f}{\partial t} = -\frac{1}{2} t^{-3/2} e^{-\frac{(x-y)^2}{4Dt}} + t^{-1/2}\frac{(x-y)^2}{4Dt^2}e^{-\frac{(x-y)^2}{4Dt}}
= \left[ \frac{(x-y)^2}{4Dt} - \frac{1}{2} \right] t^{-3/2} e^{-\frac{(x-y)^2}{4Dt}}
$$

Deriverer to ganger med hensyn på x på høyresiden:
$$\frac{\partial f}{\partial x} = t^{-1/2} \frac{-2(x-y)}{4Dt} e^{-\frac{(x-y)^2}{4Dt}} = -\frac{1}{2D} t^{-3/2} \left[ (x-y) e^{-\frac{(x-y)^2}{4Dt}} \right]$$

$$\frac{\partial^2 f}{\partial x^2} = -\frac{1}{2D} t^{-3/2} \left[ e^{-\frac{(x-y)^2}{4Dt}} - \frac{2(x-y)^2}{4Dt} e^{-\frac{(x-y)^2}{4Dt}}\right] = \frac{1}{D} \left[ \frac{(x-y)^2}{4Dt} - \frac{1}{2} \right] t^{-3/2} e^{-\frac{(x-y)^2}{4Dt}}$$

Av dette ser vi lett at $\frac{\partial f}{\partial t} = D\frac{\partial^2 f}{\partial x^2}$, som var det vi ville vise.

Fysisk tolkning:

Hvis vi lar $t \rightarrow 0$ i $\phi(x,t)$ ser vi at eksponensialleddet i integralet går mot 0 i alle andre punkter enn $x = y$. Vi kan tolke dette som at alle partiklene starter samlet i punktet $x = y$ ved $t = 0$. Det må bety at funksjonen $h(y)$ på et vis beskriver hvordan spredningen av partiklene skjer avhengig av startposisjonen.

Fra formelen for integralet til en generell gaussfunksjon $\int_{-\infty}^{\infty} e^{-a(x+b)^2} \mathrm{d}x = \sqrt{\frac{\pi}{a}}$ Ser vi at funksjonen vår forenkler seg til et normert integral med verdi konstant lik C når $h(y) = C$. For funksjoner som avhenger av y vil vi få mer kompliserte fordelinger.

**Vent dette er fakka:** Vi integrerer jo bort y, så fordelingen er ikke avhengig av hvor vi starter? Men det vil jo det? Vi kan jo ikke evaluere integralet før vi har en y og en t. Er jo derfor vi har integralet der i utgangspunktet. Vi kan ikke evaluere det før vi har en x og en t.

#### Oppgave 1b)

Vi har at $P(x) = \frac{e^{-\beta V(x)}}{Z}$, der $\beta = (k_BT)^{-1}$. Vi lar sannsynligheten for at an partikkel beveger seg fra en posisjon x til $x_0 \in \{x-1, x, x+1\}$ er proposjonal med $P(x)$, og vi antar at denne proposjonalitetskonstanten er lik for alle mulige steg. VIl bestemme sannsynlighetene $p^+, p^0, p^-$ for at partikkelen flytter seg til høyre, blir eller til venstre. 

Vi har først kravet om at summen av de tre sannsynligghetene skal være 1 (vi kan se bort fra proposjonalitetsfaktoren).

$$\frac{e^{-\beta V(x_0 + 1)}}{Z} + \frac{e^{-\beta V(x_0)}}{Z} + \frac{e^{-\beta V(x_0 - 1)}}{Z} = 1$$
$$Z = e^{-\beta V(x_0 + 1)} + e^{-\beta V(x_0)} + e^{-\beta V(x_0 - 1)}$$

Vi får dermed:

$$ p^+ = P(x_0 + 1) = \frac{e^{-\beta V(x_0 + 1)}} {e^{-\beta V(x_0 + 1)} + e^{-\beta V(x_0)} + e^{-\beta V(x_0 - 1)}} = \frac{1}{1 + e^{-\beta[V(x_0-1)-V(x_0+1)]} + e^{-\beta[V(x_0)-V(x_0+1)]}}$$

Uttrykk for $p^0$ og $p^-$ får vi helt tilsvarende ved å regne ut $P(x_0)$ og $P(x_0-1)$, så vi skriver det ikke opp her.


#### Oppgave 1c) 

Vi ser på de to grensetilfellene for $p^+$, $p^0$ og $p^-$.

Når $k_{\mathrm{B}} T \gg |V(x+1)-V(x)|$ for alle x vil $\beta |V(x+1)-V(x)| \ll 1$. Det vil si at alle eksponensialleddene går mot $e^0 = 1$, og dermed får vi at $p^+ \approx p^0 \approx p^- \approx \frac{1}{3}$

Når $k_{\mathrm{B}} T \ll |V(x+1)-V(x)|$ for alle x vil $\beta |V(x+1)-V(x)| \gg 1$ det vil si at alle eksponensialleddene i $p^-$ går mot 0, slik at $p^- \approx 1$, mens $p_0$ og $p^+$ begge har eksponensialledd som vokser, slik at $p_0 \rightarrow 0$ og $p^+ \rightarrow 0$.

Dette synes å være fornuftig, små potensialforskjeller relativt til partiklenes energi ikke burde ha noen påvirkning på deres bevegelse, mens store potensialforskjeller burde gjøre det mer sannsynlig å bevege seg til lavere poteniale. Vi kan i disse to tilfellene forenkle virrevandringsproblemet, ved i det første å neglisjere potensialforskjellene slik at virrevandringen blir helt tilfeldig, og i det andre tilnærme det neste steget som helt sikkert.

```python
import numpy as np

"""
Tror denne funksjonen skal klare alt, i alle frall frem til 4a)!!!!!
"""
def random_walk_universal(n_steps, start_pos, p_minus, p_pluss, repulsion = False, axis_wrap = None, size = 0):

    n_part = len(start_pos)
    pos = np.copy(start_pos)
    delta_pos = np.zeros(n_part)
    period = len(p_pluss)

    if (axis_wrap is not None): 
        axis_start = axis_wrap[0]
        axis_end = axis_wrap[1]
        axis_length = axis_end-axis_start


    if (not repulsion):

        for i in range(n_steps):
            step = np.random.uniform(size = n_part)
            step[step <= p_minus] = -1
            step[(step > p_minus) & (step <= 1-p_pluss)] = 0
            step[step > 1-p_pluss] = 1

            pos = pos + step
        
        delta_pos = pos - start_pos

        if (axis_wrap is not None):
            pos = (pos-axis_start)%axis_length + axis_start
    

    if (repulsion):

        for i in range(n_steps):

            random_indexes = np.arange(n_part)
            np.random.shuffle(random_indexes)
            moves = np.zeros(n_part)

            for i in random_indexes:
                step = np.random.uniform()
                if step <= p_minus[pos[i]%period]:
                    step = -1
                elif step > (1 - p_pluss[pos[i]%period]):
                    step = 1
                else:
                    step = 0

                if (axis_wrap is not None):
                    if not (pos == (pos[i] + 2*size + step)%axis_length).any():
                        moves[i] = step
                else:
                    if not (pos == (pos[i] + 2*size + step)).any():
                        moves[i] = step

            delta_pos += moves.astype(int)

            pos += moves.astype(int)

            if (axis_wrap is not None):
                pos = (pos-axis_start)%axis_length + axis_start
    
    return pos, delta_pos
```

```python
import numpy as np
from matplotlib import pyplot as plt


def p_vals(x, V, bk):
    p_pluss = 1 / (1 + np.exp(-bk*(V(x-1)-V(x+1))) + np.exp(-bk*(V(x)-V(x+1))))
    p_minus = 1 / (1 + np.exp(-bk*(V(x+1)-V(x-1))) + np.exp(-bk*(V(x)-V(x-1))))
    return np.array([p_pluss, p_minus])


"""
Funksjon som gjør n antall steg for hver posisjon i start_pos,
og returnerer en array med endelig posisjon for hver partikkel.
"""
def random_walk(n_steps, start_pos, p_minus, p_pluss):

    pos = np.copy(start_pos)

    for i in range(n_steps):
        step = np.random.uniform(size = len(pos))
        step[step <= p_minus] = -1
        step[(step > p_minus) & (step <= 1-p_pluss)] = 0
        step[step > 1-p_pluss] = 1

        pos = pos + step
    
    return pos

"""
Konstanter
"""
steps = 200
n_sim = 10
bk = 1

"""
Plasserer 101 partikler fra -50 til 50.
"""
positions = np.arange(-50, 51, 1)
n_particles = len(positions)

def V_1(x):
    return -x

def V_2(x):
    return np.ones(len(x))

"""
Sannsynligheter for V_1 og V_2
"""
p_pluss_1, p_minus_1 = p_vals(positions, V_1, bk)
p_pluss_2, p_minus_2 = p_vals(positions, V_2, bk)

"""
Simulerer, og plotter distribusjonen
"""

walk_1 = np.zeros(n_particles*n_sim)
walk_2 = np.zeros(n_particles*n_sim)

for i in range(n_sim):
    walk_1[i*n_particles:(i+1)*n_particles] = random_walk(steps, positions, p_minus_1, p_pluss_1)
    walk_2[i*n_particles:(i+1)*n_particles] = random_walk(steps, positions, p_minus_2, p_pluss_2)
    #walk_1[i*n_particles:(i+1)*n_particles] = random_walk_universal(steps, positions, p_minus_1, p_pluss_1)[0]
    #walk_2[i*n_particles:(i+1)*n_particles] = random_walk_universal(steps, positions, p_minus_2, p_pluss_2)[0]

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 2.5))

axs[0].hist(walk_1, bins = 200, density = True)
axs[0].set_title("V(x)=-kx")
axs[1].hist(walk_2, bins = 200, density = True)
axs[1].set_title("V(x)=k")

for ax in axs:
    ax.set_xlabel("Position")
    ax.set_ylabel("Particle density")
```

```python

"""
Funksjon som gjør n_steps steg for hver posisjon i start_pos med hard repulsjon.
Rekkefølgen repulsjonen evalueres i er tilfeldig.
"""
def random_walk_repulsion(n_steps, start_pos, p_minus, p_pluss):

    pos = np.copy(start_pos)
    n_part = len(start_pos)

    for i in range(n_steps):

        random_indexes = np.arange(n_part)
        np.random.shuffle(random_indexes)

        for i in random_indexes:
            step = np.random.uniform()
            if step <= p_minus[i]:
                step = -1
            elif step > (1 - p_pluss[i]):
                step = 1
            else:
                step = 0

            if not (pos == (pos[i] + step)).any():
                pos[i] += step
    
    return pos

"""
Simulerer, og plotter distribusjonen.
"""

walk_1 = np.zeros(n_particles*n_sim)
walk_2 = np.zeros(n_particles*n_sim)

for i in range(n_sim):
    walk_1[i*n_particles:(i+1)*n_particles] = random_walk_repulsion(steps, positions, p_minus_1, p_pluss_1)
    walk_2[i*n_particles:(i+1)*n_particles] = random_walk_repulsion(steps, positions, p_minus_2, p_pluss_2)
    #walk_1[i*n_particles:(i+1)*n_particles] = random_walk_universal(steps, positions, p_minus_1, p_pluss_1, repulsion = True)[0]
    #walk_2[i*n_particles:(i+1)*n_particles] = random_walk_universal(steps, positions, p_minus_2, p_pluss_2, repulsion = True)[0]

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 2.5))

axs[0].hist(walk_1, bins = 200, density = True)
axs[0].set_title("V(x)=-kx")
axs[1].hist(walk_2, bins = 200, density = True)
axs[1].set_title("V(x)=k")

for ax in axs:
    ax.set_xlabel("Position")
    ax.set_ylabel("Particle density")
```

For kreteriet med "hard repulsjon" ser vi at partiklene har liknende oppførsel, men at spredningen blir større (siden partiklene ikke kan passere forbi hverandre). En måte å tenke på det er atomer i en gass som kolliderer med hverandre, slik at det oppstår et indre "trykk" som presser partiklene utover.

Videre ser vi at når vi senker temperaturen i deloppgave c, påvirker ikke dette partiklene under konstant potenial i særlig grad. For partiklene i det synkende potensialet derimot, blir bevegelsen til høyere enda mer dominerende. I tillegg blir forskjellen i spredning med og uten hard repulsjon enda tydeligere. Partiklene lengst til høyre beveger seg omtrent like langt som de som ikke er utsatt for hard repulsjon, men partiklene lengst til venstre har beveget seg mye kortere. En analogi til dette er biler som står i kø ved et trafikklys. Å ikke ha hard repulsjon blir som at alle sjåførene reagerer og begynner å kjøre i samme hastighet samtidig. Hard repulsjon blir nærmere slik det er i virkeligheten: Man kan ikke kjøre når det er en bil bare en mater foran. I tillegg har ikke alle sjåførene samme reaksjonstid, og resultatet er at bilkolonnen strekker seg ut.

## Oppgave 3: Virrevandring i skrallepotensial uten interaksjoner

#### Oppgave 3a)

```python
"""
Funksjon som gjør n_steps steg for hver posisjon i start_pos, med periodisk potensial gitt ved lengden på p_minus og p_pluss
Rekkefølgen repulsjonen evalueres i er rekkefølgen til start_pos_vektoren.
"""
def random_walk_periodic(n_steps, x_start, p_minus, p_pluss):
    period = len(p_minus)

    pos = np.copy(x_start)

    for i in range(n_steps):
        p_vals_pluss = p_pluss[pos%period]
        p_vals_minus = p_minus[pos%period]
        step = np.random.uniform(size = len(pos))
        step[step <= p_vals_minus] = -1
        step[(step > p_vals_minus) & (step <= 1-p_vals_pluss)] = 0
        step[step > 1-p_vals_pluss] = 1

        pos += step.astype(int)
    
    return pos

alpha = 0.8
T_p = 500
N_x = 100
N_p = 12*N_x
bk = 1000
cycles = 10

def V_1(x):
    per_x = x%(N_x)-(1-alpha)*N_x
    return np.where(per_x > 0, per_x/(alpha*N_x), -per_x/((1-alpha)*N_x))

def run_cycles(cycles, start_position):
    positions = start_position.copy()

    currents = np.zeros(cycles)

    for i in range(cycles):
        start_position_cycle = positions.copy()
        positions = random_walk_periodic(T_p, positions, p_minus_1, p_pluss_1)
        positions = random_walk_periodic(T_p, positions, p_minus_2, p_pluss_2)
        
        currents[i] = np.mean(positions-start_position_cycle)/(2*T_p)
    
    return currents

grid = np.arange(0, 2*N_x)
start_pos  = np.repeat(np.arange(0, 2*N_x), N_p//(2*N_x))

p_pluss_1, p_minus_1 = p_vals(grid, V_2, bk)
p_pluss_2, p_minus_2 = p_vals(grid, V_1, bk)
currents_1 = run_cycles(cycles, start_pos)
potential_curve_1 = V_1(grid)


alpha = 0.1

p_pluss_1, p_minus_1 = p_vals(grid, V_2, bk)
p_pluss_2, p_minus_2 = p_vals(grid, V_1, bk)
currents_2 = run_cycles(cycles, start_pos)
potential_curve_2 = V_1(grid)

with np.printoptions(precision=10, suppress=True, linewidth = 100):
    print("Average currents for the 10 cycles:")
    print(f"a=0.8: {currents_1}")
    print(f"a=0.1: {currents_2}")

fig, axs = plt.subplots(nrows = 1, ncols = 2, figsize=(10, 2.5))

axs[0].plot(grid, potential_curve_1)
axs[1].plot(grid, potential_curve_2)
axs[0].set_title(r"$ \alpha = 0.8$")
axs[1].set_title(r"$ \alpha = 0.1$")
for ax in axs:
    ax.set_xlabel("x")
    ax.set_ylabel("V_1")
plt.show()
```

Tolkning av resultatene:

* Vi får en negativ strøm som følge av at i størsteparten av området er potensialet synkende mot venstre. Det medffører at partiklene strømmer mot venstre.

* La oss forsøke å forklare differansen mellom $J_{avg}(0)$ og $J_{avg}(n)$ for $n>0$. La oss starte med å anta at strømmen mens partiklene er i potensialet $V_2$ er neglisjerbar, slik at vi bare betrakter partiklenes bevegelse i $V_1$ for å forklare strømmen. I starten er alle partiklene jevnt fordelt på posisjoner fra 0 til 200, men etter de første $T_p$ stegene vil det ha oppstått en opphoping av partikler i potensialbrønnene som fører til at strømmen blir lavere i de neste syklusene. Årsaken til at strømmen ikke stopper opp helt er at mens partiklene er i potensialet $V_2$, vil noen av partiklene ha beveget seg over til andre siden av "fjelltoppen", slik at de begynner å skli ned igjen idet de kommer tilbake igjen i $V_1$.

Gjentatt med $\alpha = 0.1$:

* Ser nå at vi har helning andre vei, slik at strømmen går mot høyere. Vi ser at "potensialveggen" partiklene møter er enda brattere enn for $\alpha = 0.8$, slik at sannsynligheten er større for at partiklene kommer seg over til andre siden i løpet av tiden de er i $V_2$, slik at strømmen blir større.
